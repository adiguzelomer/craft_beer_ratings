```python
def get_beer_mega_review(brew_beer: str, reviews_df: pd.DataFrame):
    """Return a string of the reviews associated with the given beer.
    """
    beer_reviews = reviews_df[reviews_df['brew_beer']==brew_beer]
    texts = beer_reviews['cleaned_text'].values
    return ' '.join(list(texts))

get_beer_mega_review('russian river brewing company scratching post', text)
```

```python
def get_all_beer_tf_idf(brews: list, reviews_df: pd.DataFrame, tfidf_vectorizer):
    """Return the tf_idf matrix for all the beers listed.
    """
    all_reviews = [get_beer_mega_review(brew_beer, reviews_df) for brew_beer in brews]
    print(all_reviews[0:5])
    print(len(all_reviews))
    return tfidf_vectorizer.transform(all_reviews)

with open('models/1-nmf/TF-IDF-Vectorizer.pkl', 'rb') as f:
    tfidfvec = pickle.load(f)
    
tfidf = get_all_beer_tf_idf(beers_to_plot, text, tfidfvec)
```
